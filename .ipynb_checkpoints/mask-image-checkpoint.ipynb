```python
def Merge_Mask(maskes_dir,img_dir):
    alpha = 0
    meta = 1 - alpha
    gamma = 1.0
    
    img=cv2.imread(img_dir)
    n_mask=len(maskes_dir)
    if n_mask!=0:
        msk=cv2.imread(maskes_dir[0])
        imag=cv2.addWeighted(img,alpha,msk,meta,gamma)
        for i,mask_dir in enumerate(maskes_dir[1:]):
            mask=cv2.imread(mask_dir)
            im= cv2.addWeighted(img,alpha,mask,meta,gamma)
            imag=cv2.add(imag,im) 
    #imag=cv2.cvtColor(imag,cv2.COLOR_RGB2GRAY)
    return imag,img
```

```python
mask,img=Merge_Mask(maskes_dir=train_mask[10]["mask_path"],img_dir=train_mask[10]["path"])
#res=cv2.bitwise_and(img,img,mask)
plt.imshow(mask)
plt.imshow(cv2.cvtColor(mask,cv2.COLOR_RGB2GRAY))
```
